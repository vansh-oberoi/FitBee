```python
df = pd.read_csv('Final_Exercise_Recommendations_Dataset new.csv')
```

```python
df['BMIcase'].value_counts()
```

```python
gender_encoding = {'Male':0, 'Female':1}
BMIcase_encoding = {val: i[0] for i,val in np.ndenumerate(df['BMIcase'].unique())}
```

```python
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
df['BMIcase'] = df['BMIcase'].map({'Underweight': 0, 'Normal': 1, 'Overweight': 2, 'Obesity': 3})
```

```python
df['Gender'] = df['Gender'].map(gender_encoding)
df['BMIcase'] = df['BMIcase'].map(BMIcase_encoding)
```

```python
text_features = ['Recommendation Summary']  # Replace with actual text column name
categorical_features = ['Gender', 'BMIcase']
```

```python
X = df.drop(columns=['Recommendation Summary'])  # Replace 'Target' with your actual target column name
y = df['Recommendation Summary']
```

```python
y = df['Recommended Exercises']
X = df.iloc[:,0:6]
```

```python
model = CatBoostClassifier(text_features=text_features, cat_features=categorical_features, verbose=0, random_seed=42)
cv_scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')  # 5-fold cross-validation

```

```python
mp = {i[0] : val for i,val in np.ndenumerate(y.unique())}
mp2 = {val : i[0] for i,val in np.ndenumerate(y.unique())}
```

```python
y = np.array(y.map(mp2)).reshape(-1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, shuffle = True)
```

```python
model = CatBoostClassifier(verbose = 0, loss_function = 'MultiClass', n_estimators = 100, learning_rate = 0.1)
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(y_pred)
```
